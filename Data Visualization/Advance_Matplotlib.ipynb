```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Colored Scatterplots

```python
iris=pd.read_csv("iris.csv")
iris.sample(5)
```

```python
# give the iris species diffrent types to number (replace)
iris['Species'] = iris['Species'].replace({
    'Iris-setosa': 0,
    'Iris-versicolor': 1,
    'Iris-virginica': 2
})
```

```python

#we want to separte for both ('SepalLengthCm',"petalLengthCm") separate dots coolors
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel("petal Length")
plt.colorbar()

plt.show()
```

```python
# size of graph increase

plt.figure(figsize=(17,7)) #figsize=(width,height)

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel("petal Length")
plt.colorbar()

```

## Anotations

```python
batters=pd.read_csv('batter.csv')
batters.head(5)
```

```python
sample_df=batters.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]

plt.scatter(x,y)
plt.text(1,5,"point 1")
plt.text(2,6,"point 2")
plt.text(3,7,"point 3")
plt.text(4,8,"point 4",fontdict={'size':12,'color':'red'})

plt.show()
```

```python
# how can plot horizantall and vertically line

plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red') 
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
plt.show()
```

```python
# A diff  way to plot graph

batters.head()
```

```python
plt.scatter(batters['avg'],batters['strike_rate'])
```

```python
fig, ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'], batters['strike_rate'])
ax.set_title("Something")
ax.set_xlabel("Avg")
ax.set_ylabel("Strike Rate")

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))
ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg vs Strike rate')
ax[0].set_ylabel("strike rate")



ax[1].set_title('Avg vs runs')
ax[1].set_ylabel("Runs")
ax[1].set_xlabel("Avg")

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,6))
ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
fig=plt.figure()

ax1=fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2=fig.add_subplot(2,2,2) #2rows,1columns,2graph
ax2.hist(batters['runs'])

ax2=fig.add_subplot(2,2,3) #2rows,2columns,3graph
ax2.hist(batters['avg'])
```
