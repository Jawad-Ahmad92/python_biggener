## Types of Data
        1. Numerical Data (Numbers)

Definition: Data that is in the form of numbers and can be measured or calculated.

Examples:

Age = 20
Height = 175 cm
Salary = 50,000
Marks = 85

Simple words:
➡️ Numerical data is data with numbers. You can add, subtract, multiply, divide, and calculate with it.

        2. Categorical Data (Labels or Names)

Definition: Data that is in the form of names, labels, or categories instead of numbers.

Examples:

Gender = Male, Female
Color = Red, Blue, Green
City = Peshawar, Lahore, Karachi
Blood Group = A+, B+, O+

Simple words:
➡️ Categorical data is data with names or labels. It is used to group or classify things, not to calculate with them.

```python
# upload datasets

batsman=pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('RG Sharma vs virat kohli career compersion')

plt.ylabel('Runs Scored')
plt.xlabel('Season')
```

```python
# colors(hex) and line(width and style) and marker(site)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')

plt.title('RG Sharma vs virat kohli career compersion')

plt.ylabel('Runs Scored')
plt.xlabel('Season')
```

```python
# line style
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed')

plt.title('RG Sharma vs virat kohli career compersion')

plt.ylabel('Runs Scored')
plt.xlabel('Season')
```

```python
# d--> dymond
plt.plot(batsman['index'],batsman['V Kohli'],color='green',marker='d')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='d')

plt.title('RG Sharma vs virat kohli career compersion')

plt.ylabel('Runs Scored')
plt.xlabel('Season')
```

```python
# label
plt.plot(batsman['index'],batsman['V Kohli'],color='green',marker='d',label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='d',label='rohit')

plt.title('RG Sharma vs virat kohli career compersion')

plt.ylabel('Runs Scored')
plt.xlabel('Season')

plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
# stacked bar chart

plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2015']+df['2016'],label='2017')

plt.legend()
plt.show()
```

# Practice Question

```python
# line plot
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 180, 170, 200, 220]
```

```python
plt.plot(months,sales)
plt.title("Montly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# bar plot

students=["Ali","Ahmad","sara","Ayesha","Hiba"]
marks=[85,92,78,88,95]

plt.bar(students,marks,color='green')
plt.title("Marks chart")
plt.xlabel("Student Names")
plt.ylabel("Marks")
plt.grid()
plt.show()
```

```python
# scatter graph
hours = [1, 2, 3, 4, 5, 6]
marks = [40, 50, 60, 70, 85, 95]

plt.scatter(hours,marks,color='red')
plt.title("Study Hours vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("marks")
plt.show()
```

```python
# line/plot graph

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [100, 120, 140, 130, 170, 190]
profit = [20, 25, 30, 28, 40, 45]

plt.plot(months,sales,marker='o',label="Sales")
plt.plot(months,profit,marker='s',label="profit")
plt.title("Monthly Sales and Profit")
plt.xlabel('months')
plt.ylabel("Amount")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# pie chart

languages = ["Python", "Java", "C++", "JavaScript"]
students = [40, 25, 20, 15]

plt.pie(students,labels=languages,autopct="%1.1f%%",startangle=90)

plt.xlabel("languages")
plt.ylabel("students")
plt.title("Students Favorite Programming Language")
plt.show()
```

```python
# scatter plot

hours_studied = [1, 2, 3, 4, 5, 6, 7]
marks = [45, 50, 60, 65, 72, 85, 90]

plt.scatter(hours_studied,marks,marker='s')
plt.xlabel("hours_studied")
plt.ylabel("marks")
plt.title("studeies hourse")
plt.show()
```

```python
# plot/line graph

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperature = [30, 32, 35, 34, 36, 38, 37]

plt.plot(days,temperature,color='red',linestyle='dashed',marker='*',label='Temperature')
plt.grid(True)
plt.legend()
plt.title("Daily Temperature")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```
