# ***3D Graphs***

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
batters=pd.read_csv('batter.csv')
batters.head()
```

## 3D Scatter plot

```python


fig=plt.figure(figsize=(17,7))
ax=plt.subplot(projection='3d')

ax.scatter3D(batters['runs'],batters['avg'],batters['strike_rate'],marker='+')

ax.set_title('ipl batsman analysis')

ax.set_xlabel("Runs")
ax.set_ylabel("Avg")
ax.set_zlabel("strike rate")
```

## 3D Line plot

```python
x=[0,1,5,30]
y=[0,10,13,50]
z=[0,13,20,43]

fig=plt.figure()
ax=plt.subplot(projection='3d')

ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='red')
```

## 3D Surface Plots

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)

z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')

p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')

p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## contour graph

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()

p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()

p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Heatmap

```python
delivery=pd.read_csv("IPL_Ball_by_Ball_2008_2022.csv")
delivery.head()
```

```python
temp_df = delivery[
    (delivery['ballnumber'].isin([1,2,3,4,5,6])) &
    (delivery['batsman_run'] == 6)
]
```

```python
grid=temp_df.pivot_table(
    index='overs',
    columns='ballnumber',
    values='batsman_run',
    aggfunc='count'
)
grid
```

```python
plt.figure(figsize=(10,10))
plt.imshow(grid)
plt.colorbar()
```
