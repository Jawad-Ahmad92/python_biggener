```python
# Set of function -->
#     1= np.union1d,
#     2= np.intersect1d,
#     3= np.setdiff1d,
#     4= np.setxor1d,
#     5= np.in1d
    
n=np.array([1,2,3,4,5])
m=np.array([3,4,5,6,7,8])
```

```python
# union

np.union1d(n,m)
```

```python
# intersect

np.intersect1d(n,m)
```

```python
# setdiff1d

np.setdiff1d(n,m)
```

```python
# setxor1d
np.setxor1d(n,m)
```

```python

# Create values of t
t = np.linspace(0, 20*np.pi, 1000)

# Spiral equations
x = t * np.cos(t)
y = t * np.sin(t)

# Plot graph
plt.plot(x, y)

# Add title and labels
plt.title("Spiral Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add grid
plt.grid(True)

# Show graph
plt.show()
```
