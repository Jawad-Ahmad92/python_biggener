ARRAY operation

```python
b1=np.arange(12).reshape(3,4)
b2=np.arange(12,24).reshape(3,4)

b2
```

scalar operatration

```python
# scalar opeario--> single value * by matrix of all element

b1*2
```

```python
# RELATONAL 
b2>15
```

 vector operation
 

```python
b1+b2 # if both matrix shap are same 
b1*b2
```

max/min/sum/prod(multiplication of all numbers)

```python
np.max(b1)
np.min(b2)
np.sum(b1)
np.prod(b2)
```

```python
# if we want min or max a specific column or rows so we used 
# 0 -->column
#  1 --> rows

np.max(b2 ,axis=1)

# output -->  array([15, 19, 23]) it mean in first rows max valuve is 15 in 2 rows max is 19 in third is 23

np.min(b1,axis=0)
```

mean / medium/ std/var

```python
np.var(b1, axis=1)
```
