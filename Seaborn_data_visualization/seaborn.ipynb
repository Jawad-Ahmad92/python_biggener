```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px  # now it's for dataset
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
# scatter plot--->x_axis level  function

sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')

plt.show()
```

Important Points for Exam  
✅ scatterplot() is an Axes-level function   
✅ Each dot represents one row of data  
✅ data → Selects the dataset  
✅ x → Horizontal axis  
✅ y → Vertical axis  
✅ hue → Changes dot color  
✅ style → Changes dot shape  
✅ size → Changes dot size  
✅ plt.show() → Displays the graph  

```python
# figure level function

sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='line')
```

## line plot 

```python
gap=px.data.gapminder()
gap.sample(100)
```

```python
# line plot

temp_df=gap[gap['country']=='Pakistan']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
# using relpot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
# hue --> style

temp_df=gap[gap['country'].isin(['Brazil','Pakistan','China'])]
temp_df
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue="country",style='continent')
```

```python
# facet plot-->side by side plot diff sex

sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',hue='sex')
```

***facet:*** they only work on figur level funtion--> work with relplot  
not work on scatterplot and lineplot function

```python
# facet plot-->side by side plot diff sex

sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',row='sex',hue='sex')
```

```python
# col wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```
