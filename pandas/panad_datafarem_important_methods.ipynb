- ***values_counts()*** they work on both (series and dataframe) but in **series** they tell us the holl series how many duplicated values are there and whicch on and other side on **dataframe** they work o holl row they tell how many rows are dublicated

```python
# series
a=pd.Series([1,1,1,2,3,3,3,2,1])
a
```

```python
# dataframe 
marks=pd.DataFrame([
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,70,14],
    [80,70,14]
],columns=['iq','marks','packege'])

marks.value_counts()
```

```python
ipl=pd.read_csv('ipl-matches.csv')
ipl.head(5)
```

```python
# find which player has won most potm --> in finals and qualifiers

ipl['MatchNumber'].str.isdigit()
```

```python
# toss decsion plot
import matplotlib.pyplot as plt
ipl['TossDecision'].value_counts().plot(kind='pie')
```

```python
# how many matches each team has played
(ipl['Team2'].value_counts() + ipl['Team1'].value_counts()).sort_values(ascending=False)
```

```python
student={
    'name':['nitish','ankit','rupesh','nan','mrityunjay','nan','rishabh','nan','aditya','nan'],
    'college':['bit','lit','vit','nan','nan','visi','ssit','nan','nan','git'],
    'branch':['eee','it','cse','nan','me','ce','civ','cse','bio','nan'],
    'cgpa':[6.66,2.25,6.41,'nan',5.60,9.00,7.40,10.00,7.40,'nan'],
    'package':[4.0,5.0,6.0,'nan',6.0,7.0,8.0,9.0,'nan','nan']
}
student=pd.DataFrame(student)
student = student.replace('nan', np.nan)
student
```

```python
student.sort_values('name',na_position='first')
```

- **dropna( series and datframe)**  it is used for remove/clean missing values

```python
# on series
student['name'].dropna()
```

```python
# on dataframe

student.dropna()
```
