```python
import numpy as np
import pandas as pd
```

Series

```python
# string

country=['india','pakistan','USA','Nepal','Srilanka']

pd.Series(country)
```

```python
# integer

runs=[31,24,78,100,66]

pd.Series(runs)
```

```python
# custom index

marks=[67,89,56,100]
subjects=["math","english","sience","urdu"]

pd.Series(marks,index=subjects)
```

```python
# setting name
pd.Series(marks,index=subjects, name="jawad marks is :")
```

Series from dictnarry

```python
marks={
    'maths':67,
    'english':58,
    'science':97,
    'urdu':68,
}

marks_series=pd.Series(marks,name="jawad marks:")

marks_series
```
