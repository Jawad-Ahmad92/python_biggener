# Timestamp object
- **Important exam points**
. pd.Timestamp() creates a Timestamp object.  
pd.to_datetime() converts strings to Timestamp objects.  
A Timestamp stores both date and time.  
You can access **.year, .month, .day, .day_name(),** and perform date arithmetic with it.  

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# creating timestamp(year/month/date)
pd.Timestamp('2026/7/19')
```

```python
# only year
pd.Timestamp('2026')
```

```python
# using text
pd.Timestamp("19 july 2026")
```

```python
# providing time also

pd.Timestamp('19 july 2026 2:27 AM')
```

```python
# using datetime.datetime onj
import datetime as dt

x=pd.Timestamp(dt.datetime(2026,7,19,2,35,56))
x
```

```python
# fetching attributes
x.year #2026
x.month #7
x.day #19
x.hour #2
x.minute #35
x.second #56
x # 2026-07-19 02:35:56
```

## to_datetime function
converts an existing obj to pandas timestamp/detetimeindex obj

```python
# simple series example
s=pd.Series(['2026/1/1','2020/1/1','2019/3/23'])
pd.to_datetime(s)
```
